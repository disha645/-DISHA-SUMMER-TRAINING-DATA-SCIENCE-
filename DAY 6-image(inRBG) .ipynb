```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img=Image.open(r"C:\Users\disha\Downloads\image 1.avif")
img=np.array(img)
h,w,c=img.shape
m_y=h//2
m_x=w//2
img[0:m_y, 0:m_x, 0] = np.clip(img[0:m_y, 0:m_x, 0] + 100, 0, 255)       # Top-Left: Red
img[0:m_y, m_x:w, 1] = np.clip(img[0:m_y, m_x:w, 1] + 100, 0, 255)       # Top-Right: Green
img[m_y:h, 0:m_x, 2] = np.clip(img[m_y:h, 0:m_x, 2] + 100, 0, 255)       # Bottom-Left: Blue
img[m_y:h, m_x:w, :2] = np.clip(img[m_y:h, m_x:w, :2] + 80, 0, 255)      # Bottom-Right: Yellow

# Display
plt.imshow(img)
plt.axis("off")
plt.show()
```
